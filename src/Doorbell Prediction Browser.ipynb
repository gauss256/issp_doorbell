```python
%matplotlib inline
import csv
import glob
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import misc
from ipywidgets import interact
```

```python
# Find the most recent submission file
summary_dir = os.path.expanduser('~/Desktop/Doorbell/src/summaries/')
submission_files = sorted(glob.glob(os.path.join(summary_dir,'*.csv')))
submission_file = submission_files[-1]
```

```python
# Read the submission entries into a list
with open(submission_file, 'rb') as f:
    predictions = np.array(list(csv.reader(f)))
predictions = predictions[1:]
predictions[0]
image_names = predictions[:,0]
p0 = predictions[:,1].astype(float)
p1 = predictions[:,2].astype(float)
```

```python
image_dir = os.path.expanduser('~/Desktop/Doorbell/images/test')
```

```python
def browse_images(tests):
    n = len(tests)
    def view_image(i):
        im = misc.imread(os.path.join(image_dir, tests[i]))
        plt.imshow(im, interpolation='nearest')
        plt.title('Training: 0: {:.3}, 1: {:.3}'.format(p0[i], p1[i]))
        plt.show()
    interact(view_image, i=(0,n-1))
```

```python
browse_images(image_names)
```
